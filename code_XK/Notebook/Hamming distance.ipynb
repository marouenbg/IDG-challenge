```python
from Bio import AlignIO
from skbio import Protein
import numpy as np
import pandas as pd
from paths import *

# alignment : https://www.genome.jp/tools-bin/clustalw

input_file = path_out_data+"/protein_alignment/Kinase_domain.fasta"
input_file = path_out_data+"/protein_alignment/clustalw.aln"
alignment = AlignIO.read( open(input_file) , "clustal" )
```

```python
dists  = np.empty([len(alignment), len(alignment)])

for i in range(len(alignment)):
    seq1 = alignment[i].seq
    p1 = Protein("".join(seq1))
    
    for j in range(0, len(alignment)):
        seq2 = alignment[j].seq
        p2 = Protein("".join(seq2))
        d = p1.distance(p2)
        dists[i,j] = d
```

```python
df_dist = pd.DataFrame(dists)
df_dist.index = [seq.description for seq in alignment]
df_dist.columns = df_dist.index
```
